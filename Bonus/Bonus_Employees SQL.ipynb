```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
engine = create_engine(f'postgresql://postgres:123@localhost:5432/EmployeeSQL')
conn = engine.connect()
```

```python
data = pd.read_sql("SELECT e.emp_no, t.title, s.salary \
                   FROM salaries as s\
                   INNER JOIN titles as t\
                   ON (s.emp_no = t.emp_no)\
                   INNER JOIN employees as e\
                   ON (s.emp_no = e.emp_no)", conn)
```

## Create a bar chart of average salary by title.

```python
avg_salary=pd.DataFrame(data.groupby("title").mean()["salary"]).reset_index()
```

```python
plt.figure(figsize = (8, 5))
x_axis = np.arange(len(avg_salary.title))
y_axis = avg_salary.salary
plt.bar(x_axis,y_axis,color='r', alpha=0.5, align="center")

plt.xticks([lable for lable in x_axis], avg_salary.title,rotation = 25)
plt.title("Average Salary for Each Title")
plt.xlabel("Titles")
plt.ylabel("Average salaries")
```
